```python
import pandas as pd

# ==========================
# EXTRACT
# ==========================

print("Iniciando extração dos dados...")

# Lê o arquivo CSV
clientes = pd.read_csv("clientes.csv")

print(f"{len(clientes)} clientes carregados com sucesso!\n")
```

```python

# ==========================
# TRANSFORM
# ==========================


print("Gerando mensagens personalizadas...\n")


def gerar_mensagem(nome):
    mensagens = {
        "Ana Silva": "Olá, Ana! Que seu planejamento financeiro deste mês te aproxime ainda mais dos seus objetivos. Conte conosco nessa jornada!",

        "Carlos Souza": "Carlos, investir com constância faz toda a diferença. Continue construindo um futuro financeiro sólido!",

        "Mariana Lima": "Mariana, pequenas decisões hoje podem gerar grandes conquistas amanhã. Continue cuidando do seu patrimônio!",

        "Pedro Santos": "Pedro, organização financeira é liberdade. Continue acompanhando seus investimentos e evoluindo todos os dias!",

        "Juliana Alves": "Juliana, cada passo rumo aos seus objetivos financeiros conta. Estamos aqui para ajudar nessa caminhada!"
    }

    return mensagens.get(
        nome,
        f"Olá, {nome}! Obrigado por fazer parte da nossa instituição."
    )


# Cria a nova coluna com as mensagens
clientes["mensagem"] = clientes["nome"].apply(gerar_mensagem)


# Exibe as mensagens geradas
print("Mensagens geradas:\n")

for _, cliente in clientes.iterrows():
    print(f"Cliente: {cliente['nome']}")
    print(f"Mensagem: {cliente['mensagem']}")
    print("-" * 70)
```

```python
# ==========================
# LOAD
# ==========================


clientes.to_csv(
    "clientes_processados.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nProcesso ETL concluído com sucesso!")
print("Arquivo salvo em: data/clientes_processados.csv")
```
